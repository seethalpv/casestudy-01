Case study on Presentation of data

Do the following operations on ‘cars_data’ dataset:
1. Read the dataset to the python environment.
2. Check for the null values present in the dataset.
3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Read the dataset to the python environment.

```python
data=pd.read_csv('cars_data.csv')
data.head()
```

2. Check for the null values present in the dataset.

```python
data.isnull().sum()
```

3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(10,5))
plt.bar(data['Buyer Gender'],data['Sale Price'],width=0.7)
plt.title('male vs female buyers')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Male Vs Female buyers')
sns.countplot(data['Buyer Gender'])
plt.minorticks_on()
plt.grid(color='green')
plt.grid(color='blue')
plt.show()
```

4. Find the top 5 cars based on their sales price.

```python
cardata=data.sort_values(by=['Sale Price'], ascending=False)
cardata.head(5)
```

5. Find the least 5 cars based on their Resell price.

```python
cars = data.sort_values(by=['Resell Price'], ascending=True)
cars.head(5)
```
